```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
grid = gpd.read_file('./0.5 Grid/0.5.shp')
data = gpd.read_file('./final_data.gpkg')
```

```python
data['row_index'] = (data['LONGITUDE']-111.95)//0.5
data['col_index'] = (-9.95-data['LATITUDE'])//0.5
data['grid_index'] = data['col_index'] + data['row_index']*69+1
```

```python
#以grid_index排序
data = data.sort_values(by=['grid_index'])
```

```python
#grid新设置列记录从2001年1月到2025年2月的数据
for i in range(2001, 2026):
    for j in range(1, 13):
        grid[str(i)+'-'+str(j)] = np.nan
```

```python
grid
```

```python
#将data列名中类似2001-01改为2001-1，以便与grid列名对应
for i in range(2001, 2026):
    for j in range(1, 13):
        if j < 10:
            data = data.rename(columns={str(i)+'-0'+str(j):str(i)+'-'+str(j)})
        else:
            data = data.rename(columns={str(i)+'-'+str(j):str(i)+'-'+str(j)})
```

```python
#data根据grid_index分组
grouped = data.groupby('grid_index')
```

```python
#以grid_index为分组，计算data中2001年1月到2025年2月的数据的平均值
grouped = data.groupby('grid_index')
for i in range(2001, 2025):
    for j in range(1, 13):
        grid[str(i)+'-'+str(j)] = grouped[str(i)+'-'+str(j)].mean()
grid['2025-1'] = grouped['2025-1'].mean()
grid['2025-2'] = grouped['2025-2'].mean()
```

```python
grid.to_file('./rain_data.gpkg', driver='GPKG')
```

以下是人口密度计算

```python
pop_data = gpd.read_file('./32180_ERP_2023_SA2_GDA2020/32180_ERP_2023_SA2_GDA2020.gpkg')
```

```python
pop_data.drop(columns=['ERP_change_number_2022_23',
       'ERP_change_per_cent_2022_23',
       'Pop_density_2023_people_per_km2', 'Births_2021_22', 'Deaths_2021_22',
       'Natural_increase_2021_22', 'Internal_arrivals_2021_22',
       'Internal_departures_2021_22', 'Net_internal_migration_2021_22',
       'Overseas_arrivals_2021_22', 'Overseas_departures_2021_22',
       'Net_overseas_migration_2021_22', 'Births_2022_23', 'Deaths_2022_23',
       'Natural_increase_2022_23', 'Internal_arrivals_2022_23',
       'Internal_departures_2022_23', 'Net_internal_migration_2022_23',
       'Overseas_arrivals_2022_23', 'Overseas_departures_2022_23',
       'Net_overseas_migration_2022_23', 'SA2_code_2021', 'SA2_name_2021',
       'SA3_code_2021', 'SA3_name_2021'], inplace=True)
```

```python
#计算pop_data中每个地区的人口密度
pop_dense_data = pop_data.copy()
for i in range(2001, 2024):
    pop_dense_data['ERP_'+str(i)] = pop_dense_data['ERP_'+str(i)] / pop_dense_data['Area_km2']
```

```python
pop_dense_data.head()
```

```python
grid = gpd.read_file('./0.5 Grid/0.5.shp')
pop_dense_data.to_crs(epsg=3577, inplace=True)
grid.to_crs(epsg=3577, inplace=True)
```

```python
grid["grid_area"] = grid.geometry.area
```

```python
intersection = gpd.overlay(grid, pop_dense_data, how='intersection')
```

```python

intersection.to_crs(epsg=3577, inplace=True)
grid.to_crs(epsg=3577, inplace=True)
```

```python
intersection["int_area"] = intersection.geometry.area
grid["grid_area"] = grid.geometry.area
```

```python
target_columns = []
for i in range(2001, 2024):
    intersection["ERP_dense"+str(i)] = intersection["ERP_"+str(i)] * (intersection["int_area"] / intersection["grid_area"])
    target_columns.append("ERP_dense"+str(i))
weighted_density = intersection.groupby("id")[target_columns].sum().reset_index()
    
 
```

```python
#检测
g_area_1 = intersection[intersection['id'] == 4948]['int_area'].sum()
g_area_2 = intersection[intersection['id'] == 4948]['grid_area'].mean()
g_area_1/g_area_2
```

```python
grid = grid.merge(weighted_density, on="id", how="left")
```

```python
grid.to_file('./pop_dense_data.gpkg', driver='GPKG')
```
